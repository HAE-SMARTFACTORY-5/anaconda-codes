## 3. 이미지를 Server에 업로드 (Response로 ImageURL 받음)

```python
# 코랩 테스트용 이미지 업로드
uploaded = files.upload()  # 사용자에게 파일 업로드 창을 띄움
```

```python
def saveImage(filePath, fileName):
  # multipart/form-data 형식으로 요청
  files = {
      "file": (fileName, open(filePath, "rb"), "image/*")
  }

  # API Request
  print("POST Request To POST /images/upload")
  response = requests.post(f"{BASE_URL}/images/upload", files=files)

  # API Response
  print("📌 POST /images/upload Response:", response.status_code)
  print(response.json())
  return response.json()['imagePath'] # imagePath 반환

# 수정해주세요~~
# *더미* 업로드할 이미지 파일 경로
filePath = "/content/cover.png"
fileName = "cover.png"
# *더미*

imagePath = saveImage(filePath, fileName)
```

## 4. 종이컵 데이터를 Server에 저장 (이미지 URL, 에러타입들)
    - 에러타입은 이번 프로젝트에서는 Color Hex 값.
    - 에러타입은 list 타입, 에러타입이 없다면 None로 전달.

```python
def savePapercup(imagePath, errorTypes):
    # Request Body
    data = {
      "imageUrl": f"{BASE_URL}/{imagePath}",
      "errorType": errorTypes
    }

    # API Request
    print("POST Request To POST /papercups")
    response = requests.post(f"{BASE_URL}/papercups", json=data)

    # API Response
    print("📌 POST /papercups Response:", response.status_code)
    print(response.json())

# 수정해주세요~~
# *더미* 결함 탐지 (이번에는 color 값)
errorTypes = ["red", "yellow"]
# errorType이 없을 경우 None
# errorTypes = None
# *더미*

# 종이컵 데이터 업로드
savePapercup(imagePath, errorTypes)
```
