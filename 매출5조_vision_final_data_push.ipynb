```python
def saveImage(filePath, fileName):
  # multipart/form-data í˜•ì‹ìœ¼ë¡œ ìš”ì²­
  files = {
      "file": (fileName, open(filePath, "rb"), "image/*")
  }

  # API Request
  print("POST Request To POST /images/upload")
  response = requests.post(f"{BASE_URL}/images/upload", files=files)

  # API Response
  print("ğŸ“Œ POST /images/upload Response:", response.status_code)
  print(response.json())
  return response.json()['imagePath'] # imagePath ë°˜í™˜

# ìˆ˜ì •í•´ì£¼ì„¸ìš”~~
# *ë”ë¯¸* ì—…ë¡œë“œí•  ì´ë¯¸ì§€ íŒŒì¼ ê²½ë¡œ
filePath = "/content/cover.png"
fileName = "cover.png"
# *ë”ë¯¸*

imagePath = saveImage(filePath, fileName)
```

```python
def savePapercup(imagePath, errorTypes):
    # Request Body
    data = {
      "imageUrl": f"{BASE_URL}/{imagePath}",
      "errorType": errorTypes
    }

    # API Request
    print("POST Request To POST /papercups")
    response = requests.post(f"{BASE_URL}/papercups", json=data)

    # API Response
    print("ğŸ“Œ POST /papercups Response:", response.status_code)
    print(response.json())

# ìˆ˜ì •í•´ì£¼ì„¸ìš”~~
# *ë”ë¯¸* ê²°í•¨ íƒì§€ (ì´ë²ˆì—ëŠ” color ê°’)
errorTypes = ["red", "yellow"]
# errorTypeì´ ì—†ì„ ê²½ìš° None
# errorTypes = None
# *ë”ë¯¸*

# ì¢…ì´ì»µ ë°ì´í„° ì—…ë¡œë“œ
savePapercup(imagePath, errorTypes)
```
